```python
# importing NSE library
from nsetools import Nse
```

```python
# importing pandas and numpy 
import pandas as pd
import numpy as np
```

```python
nse = Nse()
```

```python
# getting information about a specfic stock
q = nse.get_quote('infy')
print(q)
```

```python
# all stock codes
nse.get_stock_codes()
```

```python
# Getting market indices list
nse.get_index_list()
```

```python
# Getting information about a market index
nse.get_index_quote('NIFTY 50')
```

```python
# top gainers dataframe
df = pd.DataFrame(nse.get_top_gainers())
df
```

```python
# top gainers dataframe
df = pd.DataFrame(nse.get_top_losers())
df
```

```python
# Getting a Stock Data in DataFrame
df = pd.DataFrame()
q = nse.get_quote('infy')
df[q['companyName']] = pd.Series(q)
# To display all rows
# pd.set_option('display.max_rows', None)
df
```
